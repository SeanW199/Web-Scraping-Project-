```python
import pandas as pd
import numpy as np
import time
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
import nltk
from nltk import word_tokenize
from nltk import pos_tag
from nltk import bigrams
from textblob import TextBlob
import operator
import requests as request
from datetime import datetime
```

```python
startPage = 1
nextPage = 0
endPage = 0
maxPages = []
link = 'https://uk.trustpilot.com/review/www.matalan.co.uk?page='
df = pd.DataFrame(columns=['Name','Title','Text','Rating', 'Date of Experience', 'Date Posted'])
```

```python
lpURL = link + str(startPage)

page = request.get(lpURL)
soupEP = BeautifulSoup(page.content, 'html.parser')

totalPages = soupEP.find_all('a', class_="link_internal__7XN06 button_button__T34Lr button_m__lq0nA button_appearance-outline__vYcdF button_squared__21GoE link_button___108l pagination-link_item__mkuN3")

for totalPages in totalPages:
    n = totalPages.find('span',class_='typography_heading-xxs__QKBS8 typography_appearance-inherit__D7XqR typography_disableResponsiveSizing__OuNP7').text
    maxPages.append(int(n))
endPage = max(maxPages)
```

```python
for x in range(endPage):
    
    nextPage = nextPage + 1
    url = link + str(nextPage)

    page = request.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')

    reviews = soup.find_all('div', class_="styles_cardWrapper__LcCPA styles_show__HUXRb styles_reviewCard__9HxJJ")

    x = x + 1
    time.sleep(3)
    
    for review in reviews:
        
        name = review.find('span', class_="typography_heading-xxs__QKBS8 typography_appearance-default__AAY17").text
        
        rating = review.find('div', class_="star-rating_starRating__4rrcf star-rating_medium__iN6Ty")
        rating = rating.find('img')
        rating = rating.attrs['alt']
    
        title = review.find('h2', class_="typography_heading-s__f7029 typography_appearance-default__AAY17").text
        
        text = review.find('p', class_="typography_body-l__KUYFJ typography_appearance-default__AAY17 typography_color-black__5LYEn")
        if text is not None:
            text = review.find('p', class_="typography_body-l__KUYFJ typography_appearance-default__AAY17 typography_color-black__5LYEn").text
        else:
            text = ""
        
        doe = review.find('p', class_="typography_body-m__xgxZ_ typography_appearance-default__AAY17 typography_color-black__5LYEn").text

        posted = review.find('time')
        posted = posted.get('datetime') 
        
        df = df.append({'Name': name,'Title': title,'Text': text,'Rating': rating, 'Date of Experience': doe, 'Date Posted': posted}, ignore_index=True)
```
