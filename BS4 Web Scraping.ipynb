```python
import pandas as pd
import numpy as np
import time
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
import nltk
from nltk import word_tokenize
from nltk import pos_tag
from nltk import bigrams
from textblob import TextBlob
import operator
import requests as request
from datetime import datetime
```

```python
startPage = 1
nextPage = 0
endPage = 0
maxPages = []
link = 'https://uk.trustpilot.com/review/www.matalan.co.uk?page='
df = pd.DataFrame(columns=['Name','Title','Text','Rating', 'Date of Experience', 'Date Posted'])
```

```python
lpURL = link + str(startPage)

page = request.get(lpURL)
soupEP = BeautifulSoup(page.content, 'html.parser')

totalPages = soupEP.find_all('a', class_="link_internal__7XN06 button_button__T34Lr button_m__lq0nA button_appearance-outline__vYcdF button_squared__21GoE link_button___108l pagination-link_item__mkuN3")

for totalPages in totalPages:
    n = totalPages.find('span',class_='typography_heading-xxs__QKBS8 typography_appearance-inherit__D7XqR typography_disableResponsiveSizing__OuNP7').text
    maxPages.append(int(n))
endPage = max(maxPages)
```

```python
for x in range(endPage):
    nextPage = nextPage + 1
    url = link + str(nextPage)
    page = request.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    reviews = soup.find_all('div', class_="styles_cardWrapper__LcCPA styles_show__HUXRb styles_reviewCard__9HxJJ")
    x = x + 1
    time.sleep(3)
    for review in reviews:
        name = review.find('span', class_="typography_heading-xxs__QKBS8 typography_appearance-default__AAY17").text
        rating = review.find('div', class_="star-rating_starRating__4rrcf star-rating_medium__iN6Ty")
        rating = rating.find('img')
        rating = rating.attrs['alt']
        title = review.find('h2', class_="typography_heading-s__f7029 typography_appearance-default__AAY17").text
        text = review.find('p', class_="typography_body-l__KUYFJ typography_appearance-default__AAY17 typography_color-black__5LYEn")
        if text is not None:
            text = review.find('p', class_="typography_body-l__KUYFJ typography_appearance-default__AAY17 typography_color-black__5LYEn").text
        else:
            text = ""
        doe = review.find('p', class_="typography_body-m__xgxZ_ typography_appearance-default__AAY17 typography_color-black__5LYEn").text
        posted = review.find('time')
        posted = posted.get('datetime') 
        
        df = df.append({'Name': name,'Title': title,'Text': text,'Rating': rating, 'Date of Experience': doe, 'Date Posted': posted}, ignore_index=True)
```

```python
def SentimentScore(reviewsDataSet):
    
    blob = TextBlob(reviewsDataSet)
    SentimentScore = 0 
    SentimentScore += blob.sentiment.polarity
    Sentiment_Score.append(SentimentScore)
    if SentimentScore > 0:
        return True
    else: 
        return False

SentimentDataset = Dataset


SentimentDataset = SentimentDataset.dropna()
positiveReviewCount = 0
negativeReviewCount = 0 
reviewsDataSet = SentimentDataset['Text']
Sentiment_Score = []
Sentiment = []


for index, review in enumerate(reviewsDataSet):

    if SentimentScore(review) == False:
        negativeReviewCount += 1
        Sentiment.append("Negative")
    else:
        Sentiment.append("Positive")
        positiveReviewCount += 1      

df1 = Dataset.assign(Sentiment_Score = Sentiment_Score)
df2 = df1.assign(Sentiment = Sentiment)
df2.loc[df2['Sentiment_Score'] == 0, 'Sentiment'] = 'Neutral'
```

```python
def compute_frequency_of_bigrams(bigrams_with_pos_tags):
    bigrams_with_frequencies = {}
    for first_element, second_element in bigrams_with_pos_tags:
        first_word_of_bigram, first_word_pos_tag = first_element[0], first_element[1]
        second_word_of_bigram, second_word_pos_tag = second_element[0], second_element[1]
        if (first_word_pos_tag == 'JJ' and second_word_pos_tag == 'NN'
           or first_word_pos_tag == 'NN' and second_word_pos_tag == 'NN'
           or first_word_pos_tag == 'NNP' and second_word_pos_tag == 'NNP'
           or first_word_pos_tag == 'NNP' and second_word_pos_tag == 'NN'):
            if (first_word_of_bigram, second_word_of_bigram) in bigrams_with_frequencies:
                bigrams_with_frequencies[(first_word_of_bigram, second_word_of_bigram)] += 1
            else:
                bigrams_with_frequencies[(first_word_of_bigram, second_word_of_bigram)] = 1
        else:
            continue
    return bigrams_with_frequencies
```

```python
def tokenBigramWF(string):
    Count = 0
    bigramList = []
    joinedList = []
    words = word_tokenize(string)
    words_with_pos_tags = pos_tag(words)
    bigrams_with_pos_tags = list(bigrams(words_with_pos_tags))
    bigrams_with_frequencies = compute_frequency_of_bigrams(bigrams_with_pos_tags)
    bigrams_with_frequencies_sorted = dict(sorted(bigrams_with_frequencies.items(),
                                                 key=operator.itemgetter(1), reverse=True)) 
    for bigram in bigrams_with_frequencies_sorted:
        if(Count <= 40):
            bigramList.append(bigram)
            Count += 1
        else:
             break   
    for x in range(len(bigramList)):
        joinedList.append(' '.join(bigramList[x]))
        
    return joinedList
```
